```python
tumulkelerpop=datanufus["Country Name"].value_counts().reset_index()["Country Name"].tolist()
del_country=[]
```

```python
for x in tumulkeler:
    if x not in tumulkelerpop:
        del_country.append(x)

del_country
```

```python
data.columns
```

```python
yeni_data=data[["Country name", "year", "Life Ladder", "Log GDP per capita",
       "Social support", "Healthy life expectancy at birth",
       "Freedom to make life choices", "Generosity",
       "Perceptions of corruption","region"]].copy()
```

```python
yeni_data.head()
```

```python
yeni_data = yeni_data[~yeni_data["Country name"].isin(del_country)]
yeni_data = yeni_data[~yeni_data.year.isin([2005, 2006, 2007, 2018, 2019, 2020])]
pop_dics = {x: {} for x in range(2008, 2018)}
for i in range(len(datanufus)):
    if datanufus["Year"][i] in range(2008, 2018):
        # Yıl ve ülke adıyla popülasyon sayısını sözlükte tutuyoruz
        pop_dics[datanufus["Year"][i]][datanufus["Country Name"][i]] = datanufus["Count"][i]
```

```python
pop_dics
```

```python
population=[]
for i in yeni_data.index:
    population.append(pop_dics[yeni_data["year"][i]][yeni_data["Country name"][i]])
yeni_data["population"]=population    
    
```

```python
fig=px.scatter(yeni_data,x="Log GDP per capita",y="Life Ladder",animation_frame="year",animation_group="Country name",
                size="population",template="plotly_white",color="region",
                hover_name="Country name",size_max=60)
fig.update_layout(title="ülkelerin gelirlerine göre mutluluk oranı")
fig.show()
```
